```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
file_path = 'data/muster.xml'
with open(file_path, 'r') as file:
    xml_content = file.read()
```

```python
soup = BeautifulSoup(xml_content, 'xml')
```

```python
boqcgtys = soup.find_all('BoQCtgy')
```

```python
gewerke = []
for boqcgy in boqcgtys:
    nested_boqcgy = boqcgy.find('BoQCtgy')
    if nested_boqcgy!=None:
        gewerk = boqcgy.find('LblTx')
        gewerke.append(boqcgy)
        print(gewerk.text.strip())
        #print(nested_boqcgy)
```

```python
data = []
for gewerk in gewerke:
    gewerk_name = gewerk.find('LblTx').text.strip()
    gewerk_nr = gewerk['RNoPart']
    print(gewerk_name)
    untergewerke = gewerk.find_all('BoQCtgy')
    #print(untergewerke)
    for untergewerk in untergewerke:
        untergewerk_name = untergewerk.find('LblTx').text.strip()
        untergewerk_nr = untergewerk['RNoPart']
        itemlists = soup.find_all('Itemlist')
        #print(itemlist)
        for itemlist in itemlists:
            items = itemlist.find_all('Item')
            #print(items)
            for item in items:
                item_id = item['ID']
                rno_part = item['RNoPart']
                qty = item.find('Qty').text
                qu = item.find('QU').text
                description = item.find('Description').text.strip()
                text = item.find('TextOutlTxt').text.strip()

                data.append({
                    #'ItemID': item_id,
                    'oz0' : gewerk_nr,
                    'oz1' : untergewerk_nr,
                    'oz3': rno_part,
                    'gewerk': gewerk_name,
                    'untergewerk' : untergewerk_name,
                    'Kurztext' : text,
                    'Qty': qty,
                    'QU': qu,
                    'Langetext': description
                    
                })
```

```python
df = pd.DataFrame(data)
```

```python
df['untergewerk'] = df['untergewerk'].apply(lambda x: x.split()[-1])
```
